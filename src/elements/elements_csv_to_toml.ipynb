```python
import csv
from pathlib import Path
from pprint import PrettyPrinter
```

```python
csv_file = Path("/home/brock/Downloads/PubChemElements_all.csv")
data = {}
keys = []
with open(csv_file, "r", encoding="utf-8") as file:
    csv_reader = csv.reader(file)
    for key in next(csv_reader):
        keys.append(key)
    
    for entry in csv_reader:
        data[entry[1]] = {}
        for index, key in enumerate(keys):
            match key:
                case "AtomicNumber":
                    new_key = "atomic_number"
                case "Symbol":
                    new_key = "symbol"
                case "Name":
                    new_key = "name"
                case "AtomicMass":
                    new_key = "atomic_mass"
                case "CPKHexColor":
                    new_key = "cpk_hex_color"
                case "ElectronConfiguration":
                    new_key = "electron_configuration"
                case "Electronegativity":
                    new_key = "electronegativity"
                case "AtomicRadius":
                    new_key = "atomic_radius"
                case "IonizationEnergy":
                    new_key = "ionization_energy"
                case "ElectronAffinity":
                    new_key = "electron_affinity"
                case "OxidationStates":
                    new_key = "oxidation_states"
                case "StandardState":
                    new_key = "standard_state"
                case "MeltingPoint":
                    new_key = "melting_point"
                case "BoilingPoint":
                    new_key = "boiling_point"
                case "Density":
                    new_key = "density"
                case "GroupBlock":
                    new_key = "group_block"
                case "YearDiscovered":
                    new_key = "year_discovered"
            try:
                data[entry[1]][new_key] = int(entry[index])
            except:
                try:
                    data[entry[1]][new_key] = float(entry[index])
                except:

                    if key == "CPKHexColor":
                        data[entry[1]][new_key] = "#"+entry[index]
                    else:
                        data[entry[1]][new_key] = entry[index]

PrettyPrinter(indent=4, sort_dicts=False).pprint(data)
```

```python
output = Path("/home/brock/Downloads/PubChemElements_all.toml")

with open(output, "w") as out:
    out.write("[[element]]\n")
    for element in data:
        dat = data[element]
        line = ""
        line += f"[element.{element}]\n"
        for key in dat:
            datum = dat[key]
            if isinstance(datum, str):
                line += f'{key:22} = "{datum}"\n'
            else:
                line += f"{key:22} = {datum}\n"
        line += "\n"

        out.write(line)
```
